## Exercice 2

Le jeu des **reines chromatiques** est une variante du célèbre problème des N reines. 
Ce dernier consiste à placer N reines sur une grille carrée de N x N cases (un échiquier) sans que deux reines soient sur la même ligne, la même colonne ni la même diagonale.
Le jeu des reines chromatiques modifie les contraintes de placement des reines en divisant l'échiquier en autant de zones colorées distinctes que son nombre de lignes (ou de colonnes).
Plus de détails sur ces zones sont données plus loin.


L'objectif du jeu consiste à placer N reines sans que deux reines s'_attaquent_, càd. sans que deux reines soient : 

- sur une même ligne,
- sur une même colonne,
- sur deux cases adjacentes par un côté **ou** par un coin, 
- dans une même zone de couleur.

Précisons quelques détails importants. 

Le nombre de cases adjacentes, par un côté ou par un coin, d'une case donnée dépend de la position de la case donnée. 
En effet :

- les quatre cases en coin de l'échiquier possèdent trois cases adjacentes,
- sinon une case sur un bord de l'échiquier en possède cinq,
- et sinon plus généralement une case "au centre" de l'échiquier en possède huit.

On notera que la contrainte d'une reine unique par diagonale du problème des N reines (classique) n'est plus demandée.

En revanche, une reine unique par zone colorée est une contrainte spécifique à la version chromatique.

Un échiquier valide pour ce jeu satisfait les propriétés suivantes.

- Il possède exactement `nb_zones` de couleurs différentes et `nb_zones` est égal au nombre de ses lignes (et de ses colonnes).
- Chaque couleur forme une zone de cellules adjacentes par au moins un côté **et non par un coin**.
- Une couleur ne peut apparaître que dans une unique zone de cellules (telle que définie ci-dessus).

### Q4. `ligne_colonne_libre()`

Avant de placer une reine, il faut s'assurer qu'elle ne soit pas attaquée. 
Cette fonction et les deux suivantes répondent à cet objectif.


Écrire la fonction `ligne_colonne_libre()` qui vérifie si le placement d'une reine sur la case `(ligne, colonne)` ne peut pas être attaqué en ligne ou en colonne par une autre reine.

```python
def ligne_colonne_libre(grille : list[list[str]], n: int, ligne : int, colonne : int) -> bool :
    '''Verifie si la ligne colonne donne sont libre pour placer une reine'''
    #verification ligne et colonne
    for i in range(n):
        if (ord(grille[i][colonne])>=65) and (ord(grille[i][colonne])<=90):
            return False
        if (ord(grille[ligne][i])>=65) and (ord(grille[ligne][colonne])<=90):
            return False
    return True
        
```

```python
def couleur_libre(grille : list[list[str]], n: int, ligne : int, colonne : int) -> bool :
    '''Verifie si il y a pas une reine de la meme couleur place sur la ligne et la colonne'''
    lettre=0
    for i in range(n):
        for j in range(n):
            if (ord(grille[i][colonne])>=65) and (ord(grille[i][colonne])<=90):
                verif=True
                lettre=grille[i][j]
    
    if (lettre == grille[ligne][colonne]) and verif==True:
        return False
    return True
```
