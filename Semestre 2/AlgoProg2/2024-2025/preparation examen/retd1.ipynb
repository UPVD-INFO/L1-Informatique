### $\blacksquare$ Exercice. (factorielle)

La valeur de factorielle (de) $n$, où $n$ est un entier positif, est :
$n ! = 1 \times 2 \times \cdots \ \times (n-1) \times n$. On choisit
$0 ! = 1$.

2.  Écrire la fonction `factorielle()` qui calcule et retourne $n!$.
3.  La fonction `factorial()` du module `math` effectue le même calcul. Utiliser cette fonction pour définir et coder des tests unitaires de votre fonction `factorielle()`.
4.  Utiliser la fonction `factorielle()` pour écrire des fonctions qui
    calculent les quantités suivantes.

    1.  Le nombre d'arrangements (listes ordonnées) de $k$ valeurs
        prises dans un ensemble de $n$ valeurs est :
        $\mathcal{A}_n^k = {n
                  !}/{(n-k)!}, (k \le n)$.
    2.  Le nombre de combinaisons (listes non ordonnées) de $k$ valeurs
        prises dans un ensemble de $n$ valeurs est :
        $\left(_k^n\right) =
                {\mathcal{A}_n^k}/{k!}$, $(k \le n)$.

````python
def factorielle(n:int) -> int:
    '''Calcul la factorielle de N de maniere iterative'''
    #corps
    res=1
    for i in range(1,n+1):
        res= res*i
    return res

def arrangement(n:int, k:int) -> int:
    '''fonction qui calcule le nombre d'arrangements d'une liste ordonnees de maniere iterative'''
    assert k<= n
    #corps
    return factorielle(n)//factorielle(n-k)

def combinaison(n:int, k:int) -> int:
    '''fonction qui calcule le nombre de combinaison d'une liste non ordonnees de maniere iterative'''
    assert k<= n
    #corps
    return arrangement(n,k)//factorielle(k)
````

````python
from math import factorial

assert factorielle(5) == 120 == factorial(5)
assert factorielle(4) == 24 == factorial(4)

assert arrangement(5,4) == 120
assert arrangement(5,2) == 20

assert combinaison(5,4) == 5
assert combinaison(5,2) == 10
````

On rappelle quelques suites et séries numériques classiques.

-   La somme des $n$ premiers entiers :
    $s_1(n) = 1 + 2 + \cdots + (n-1) + n = \sum_{k=1}^{n} k$ ;

-   La suite de Fibonacci : $u_{n+1} = u_{n}+ u_{n-1}$ et $u_0=u_1=1$.

-   Une suite arithmético-géométrique : $u_{n+1} = u_n/3 + 2$ avec $u_0=1$.

1. Écrire des fonctions qui calculent et retournent le $n$-ième terme de ces suites. 
2. Écrire des tests unitaires adaptés. Justifier vos choix.
3.  Pour chacune de ces suites et en utilisant ces fonctions, lire une valeur de $n$ entrée au clavier, calcule le $n$-ième terme de la suite puis l'afficher à l'écran.


````python
def somme(n: int) -> int:
    '''Fait la somme des entiers partant de 1 jusqu'a n'''
    #corps
    res=0
    for i in range(n+1):
        res+=i
    return res

def fibo(n: int) -> int:
    '''Calcule la suite de fibonacci de n'''
    #corps
    if n == 0 or n == 1:
        return 1
    
    n0= 1
    n1= 1
    n2= n1+n0
    for i in range(2,n):
        n0=n1
        n1=n2
        n2= n1+n0
    return n2

def aritgeom(n: int) -> float:
    '''calcule la suite arithmetico geometrique u(n)'''
    #corps
    res=1
    for i in range(n):
        res= (res/3)+2
    #on arrondie a 10^-3
    res*= -1000
    res= res//-1
    res= res/1000
    return res
    
````

L'ASCII est un codage de caractères qui

chiffres, lettres, symboles mathématiques et de ponctuation. 

En Python,
la fonction `ord()` retourne le code ASCII d'un caractère fourni en
argument. La fonction `chr()` retourne le caractère associé à un code

1. Ecrire un programme qui affiche les 128 codes ASCII selon 16 lignes de 8 colonnes.

````python
def code_ascii_melange() -> None:
    '''Affiche le code ascii avec des E-S en 16 lignes et 8 colonnes'''
    space=0
    for i in range(128):
        print(chr(i),end=" ")
        space+=1
        if space==8:
            space=0
            print()

    return None

def code_ascii() -> list[int]:
    '''Affiche le code ascii sans E-S en 16 lignes et 8 colonnes'''
    tab=[0 for i in range(128)]
    for i in range(128):
        tab[i] = chr(i)

    return tab

def affichage_ascii(tab:list[int]) -> None:
    '''Affiche le codage ascii'''
    space=0
    for i in range(128):
        print(tab[i],end=" ")
        space+=1
        if space==8:
            space=0
            print()
    
    return None
````

2. Accompagner votre développement avec quelques tests unitaires bien choisis.
3. Proposer des solutions itératives alternatives (`for` vs. `while`) lorsque qu'elles conduisent à des traitements différents.


````python
def tabup(tab:list[int],n:int) -> bool:
    '''Verifie et renvoie si un tableau est ordonnee ou non avec For'''
    assert n==len(tab)
    #corps
    for i in range(n-1):
        if tab[i]> tab[i+1]:
            return False
    return True

def tabup_while(tab:list[int],n:int) -> bool:
    '''Verifie et renvoie si un tableau est ordonnee ou non avec while'''
    assert n==len(tab)
    #corps
    i=1
    while i<n:
        if tab[i-1]> tab[i]:
            return False
        i+=1
    return True
````

````python
def max_tab(tab:list[int],n:int) -> int:
    '''Calcul et renvoie la valeur la plus grande'''
    assert n==len(tab)
    #corps
    max= tab[0]
    for i in range(n):
        if max <= tab[i]:
            max=tab[i]
    return max

def max_ind_tab(tab:list[int],n:int) -> int:
    '''Calcul et renvoie la valeur la plus grande du dernier indice'''
    assert n==len(tab)
    #corps
    max= tab[n-1]
    ind=n-1
    for i in range(n-1,-1,-1):
        if max < tab[i]:
            max=tab[i]
            ind=i
    return ind

def min_ind_tab(tab:list[int],n:int) -> int:
    '''Calcul et renvoie la valeur la plus grande du plus petit indice'''
    assert n==len(tab)
    #corps
    max= tab[0]
    ind=0
    for i in range(n):
        if max < tab[i]:
            max=tab[i]
            ind=i
    return ind
````

### $\blacksquare$ Exercice (palindrome)

Un _palindrome_ est un mot, ou un groupe de mots, dont l'ordre des lettres
reste le même qu'on le lise de la droite vers la gauche ou inversement.
Des exemples bien connus sont "été", "kayak", "mon nom", "élu par cette
crapule". Ce dernier permet d'illustrer qu'on ne tient pas compte en
général des accents, trémas, cédilles, ni des espaces. Dans cet exercice
:

-   un mot ou un groupe de mots est représenté par une chaîne de
    caractères (`str`),

-   ces caractères sont sans accent, tréma, ni cédille : "ete"

-   les espaces sont considérés comme des caractères : ainsi "elu par
    cette crapule" n'est pas un palindrome ici.

1.  Ecrire une fonction qui teste si un argument est ou non un
    palindrome et retourne le booléen correspondant.
2. Valider votre fonction avec des tests unitaires.

````python
def palindrome(charac:str, n:int) -> bool:
    '''Verifie si un mot est un palindrome ou non'''
    assert n == len(charac)
    #corps
    for i in range(n//2):
        if charac[i] != charac[n-1-i]:
            return False
    return True
````

````python
assert palindrome("palindrome",10) == False
assert palindrome("ete",3) == True
assert palindrome("elu par cette crapule",21) == False
assert palindrome("kayak",5) == True
assert palindrome("",0) == True
````

Ecrire les fonctions qui vérifie l'orthogonalité ou non de vecteurs 

1. de taille 3,
2. de taille arbitraire $n$.


````python
def sont_orthogonaux_3D(t1: list[int],m:int, t2:list[int], n: int) -> bool:
    '''Verifie si 2 vecteurs de taille 3 sont orthogonaux ou non'''
    assert n== len(t1) and n== len(t2)
    #corps
    res= 0
    for i in range(n):
        res= res + t1[i]*t2[i]
    if res != 0:
        return False
    return True

def sont_orthogonaux_nD(t1:list[int], m:int, t2:list[int], n:int) -> bool:
    '''Verifie si 2 vecteurs de taille n sont orthogonaux ou non'''
    assert m== len(t1) and n== len(t2)
    assert n==m
    #corps
    res= 0
    for i in range(n):
        res= res + t1[i]*t2[i]
    if res != 0:
        return False
    return True
````

Après traitement, la case $(i,j)$ de ce tableau contient le résultat de $i \times j$. 
La table des multiplications entre 1 et 10 est obtenue pour $a = 1$ et $b= 10$.
1.  Utiliser cette fonction pour afficher cette table pour des valeurs de `a` et `b` entrées au clavier.

````python
def table2d(a:int, b:int) -> list[list[int]]:
    '''Calcul la table de multiplication de i*j'''
    assert a<=b
    #corps
    tab= [[0 for j in range(b-a+1)] for i in range(b-a+1)]
    #stockage des variables sur le tableau 2d
    for i in range(b-a+1):
        for j in range(b-a+1):
            tab[i][j]= (a+i)*(a+j)

    return tab

def affichage(a:int, b:int) -> None:
    '''Affiche la table de multiplication de i*j'''
    assert a<=b
    #corps

    t= table2d(a,b)
    #affichage de tableau 2d
    for i in range(b-a+1):
        for j in range(b-a+1):
            print(t[i][j],end=" ")
        print()
    
    return None
````

### $\blacksquare$ Exercice (matrices)

Ecrire les algorithmes de vérification suivants pour une matrice $M$
donnée, carrée de taille $n$ et à valeurs flottantes.

1.  $M$ est diagonale ?

2.  $M$ est symétrique ?

3.  $M$ est égale à l'identité ?

4.  $M$ est l'inverse d'une autre matrice donnée $N$ ?

````python
def est_diagonale_rec(M: list[list[int]], n: int, i: int = 0, j: int = 0) -> bool:
    '''Vérifie récursivement si une matrice donnée est diagonale'''
    if i == n:  # On a parcouru toute la matrice
        return True
    if j == n:  # Fin de ligne, on passe à la suivante
        return est_diagonale_rec(M, n, i + 1, 0)
    if (i != j and M[i][j] != 0) or (i == j and M[i][i] == 0):  # Condition d'échec
        return False
    return est_diagonale_rec(M, n, i, j + 1)  # Passage au prochain élément

def est_symetrique_rec(M:list[list[int]],n: int, i:int ,j:int) -> bool:
    if i==n:
        return True
    if j==n:
        return est_symetrique_rec(M,n,i+1,0)
    if M[i][j] != M[j][i]:
        return False
    return est_symetrique_rec(M,n,i,j+1)
````

````python
def est_diagonale(M: list[list[int]], n: int) -> bool:
    '''Verifie si une matrice donnee est diagonale'''
    assert n== len(M)
    for i in range(n):
            for j in range(n):
               if  M[i][j] != 0 and i!=j:
                    return False
               if M[i][i] == 0:
                    return False
    return True

def est_symetrique(M: list[list[int]], n: int) -> bool:
    '''Verifie si une matrice donne est symetrique'''
    assert n== len(M)

    for i in range(n):
          for j in range(n):
               if M[i][j] != M[j][i]:
                    return False
    return True

def est_identite(M: list[list[int]], n: int) -> bool:
    '''Verifie si une matrice donnee est egal a l'identite'''
    assert n== len(M)

    for i in range(n):
         for j in range(n):
               if M[i][i] != 1:
                    return False
               if M[i][j] !=0 and i!=j:
                    return False
    return True

def sont_inverse(M: list[list[int]], n1: int, N: list[list[int]],n2: int) -> bool:
    '''Verifie si une matrice donnee est diagonale'''
    assert n1== len(M)
    assert n2== len(N)
    assert n1==n2
     #creation du tableau et de son calcul
    identite= [[0 for j in range(n1)] for i in range(n1)]
    for c in range(n1):
         for i in range(n1):
              for j in range(n1):
                   identite[i][j]= M[i][j] * N[j][i]
     #verificaiton de l'identite
    for i in range(n1):
         for j in range(n1):
              if (identite[i][j]!=0 and i!=j) or (identite[i][i] != 1):
                   return False
    return True

def sont_inverse_rec(M: list[list[int]], n1: int, N: list[list[int]],n2: int,t:list[list[int]],i:int =0, j:int = 0,c:int =0) -> bool:
     if c==n1:
          return None
     elif i==n1:
          sont_inverse_rec(M,n1,N,n2,t,0,0,c+1)
     
     elif j==n1:
          sont_inverse_rec(M,n1,N,n2,t,i+1,0,c)
     else:
          t[i][j]= M[i][j] * N[j][i]
          sont_inverse_rec(M,n1,N,n2,t,i,j+1,c)
     #verification de l'est_identite
     return est_identite(t,len(t))

def sont_inverse(M: list[list[int]], n1: int, N: list[list[int]],n2: int) -> bool:
     assert n1== len(M)
     assert n2== len(N)
     assert n1==n2

     tab= [[0 for j in range(n1)] for i in range(n1)]
     return sont_inverse_rec(M,n1,N,n2,tab,0,0,0)
````
