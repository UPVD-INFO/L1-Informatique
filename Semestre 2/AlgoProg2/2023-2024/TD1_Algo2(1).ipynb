On rappelle quelques suites et séries numériques classiques.

- La somme des $n$ premiers entiers $s_1(n) = 1+ 2 + \ldots + (n+1) + n = \sum_{k=1}^n k$ ;

- La suite arithmético-géométrique $u_{n+1} = u_n/3 + 2$ avec $u_0 = 1$

- La suite de Fibonacci $u_{n+1} = u_n + u_{n-1}$ et $u_0 = u_1 = 1$


1. Pour chacune de ces suites, écrire l’algorithme qui demande et lit une valeur de $n$ entrée au clavier, calcule le $n$-ième terme de la suite puis l’affiche à l’écran.

```python
#la somme des n premiers entiers
n = int(input("n = "))

res1 = 0
for i in range(1,n+1):
    res1 = res1 + i
print(res1)

#la suite arithmético-géométrique
res2 = 1
for i in range(n):
    res2 = res2/3 + 2
print(res2)

#la suite de Fibonacci
res3 = 1
u0 = 1
u1 = 1
for i in range(1,n):
    res3 = u1 + u0
    u0 = u1
    u1 = res3
print(res3)
```

```python
def somme(n:int)->int:
    res1 = 0
    for i in range(1,n+1):
        res1 = res1 + i
    return res1

def arithmetico_geo(n:int)->float:
    res2 = 1
    for i in range(n):
        res2 = res2/3 + 2
    return res2

def fibonacci(n:int)->int:
    res3 = 1
    u0 = 1
    u1 = 1
    for i in range(1,n):
        res3 = u1 + u0
        u0 = u1
        u1 = res3
    return res3
        
print(somme(5))
print(arithmetico_geo(3))
print(fibonacci(4))
```

```python
def est_rectangle(a:float,b:float,c:float)->bool:
    tmp = 0
    if a > b:
        tmp = a
        a = b
        b = tmp
    elif b > c:
        tmp = b
        b = c
        c = tmp
    if a*a+b*b==c*c:
        return True
    return False

est_rectangle(3.0,5.0,4.0)
```

## 1.1.9. Exercice

Un palindrome est un mot, ou un groupe de mots, dont l’ordre des lettres reste le même qu’on le lise de la droite vers la gauche ou inversement. Des exemples bien connus sont “été”, “kayak”, “mon nom”, “élu par cette crapule”. Ce dernier permet d’illustrer qu’on ne tient pas compte en général des accents, trémas, cédilles, ni des espaces. Dans cet exercice :

- un mot ou un groupe de mots est représenté par une chaîne de caractères (str),

- ces caractères sont sans accent, tréma, ni cédille : “ete”

- les espaces sont considérés comme des caractères : ainsi “elu par cette crapule” n’est pas un palindrome ici.

1. Ecrire une fonction qui teste si un argument est ou non un palindrome et retourne le booléen correspondant.


```python
def palindrome(s : str, n : int) -> bool:
    for i in range(n//2):
        if s[i] != s[(n-1)-i]:
            return False
    return True
```

```python
p = "kayak"
o = "ete"
h = "elu par cette crapule"
assert palindrome(p, len(p)) == True
assert palindrome(o, len(o)) == True
assert palindrome(h, len(h)) == False
```

## Palindrome (le retour)

1. Les espaces ne doivent plus être compté

```python
def palindrome2(s : str, n : int) -> bool:
    i = 0
    j = n-1
    for k in range(n//2):
        while s[i] == " ":
            i += 1
        while s[j] == " ":
            j -= 1
            
        if s[i] != s[j]:
            return False
        
    return True
```

```python
p = "kayak"
o = "ete"
h = "elu par cette crapule"
assert palindrome2(p, len(p)) == True
assert palindrome2(o, len(o)) == True
assert palindrome2(h, len(h)) == True
```

1. $M$ est diagonale ?

```python
def diagonale(n : int, m : list[list[float]]) -> bool:
    assert n == len(m) == len(m[0])
    for i in range(n):
        for j in range(n):
            if i != j and m[i][j] != 0:
                return False
    return True
```

```python
def symetrique(n : int, m : list[list[float]]) -> bool:
    assert n == len(m) == len(m[0])
    for i in range(n):
        for j in range(i+1,n):
            if m[i][j] != m[j][i]:
                return False
    return True
```

```python
def egal_identite(n : int, m : list[list[float]]) -> bool:
    for i in range(n):
        for j in range(n):
            if i == j: 
                if m[i][j] != 1:
                    return False
            else:
                if m[i][j] != 0:
                    return False
    return True
```

4. $M$ est l’inverse d’une autre matrice donnée $N$?

```python
def produit_matrice(m : list[list[float]], n : int, o : list[list[float]], l : int):
    assert l == n
    m2 = [[0.0 for i in range(n)] for j in range(n)]
    for h in range(n):
        for i in range(n):
            for j in range(n):
                m2[h][i] += m[h][j] * o[j][i]       
    return m2

def inverse(m : list[list[float]], n : int, o : list[list[float]], l : int):
    
    C = produit_matrice(m,n,o,l)
    return egal_identite(n,C)
```

## Analyse d’une image de taille arbitraire.

1. Ecrire une fonction nbpixblc() qui compte et retourne le nombre de pixels blanc d’une image de taille arbitraire

```python
def nbpixblc(img : list[list[int]], L : int, C : int) -> int:
    """renvoie le nombre de pixels blancs dans l'image img de taille LxC"""
    cpt = 0
    for i in range(L):
        for j in range(C):
            if img[i][j] == 0:
                cpt += 1
    return cpt
```
