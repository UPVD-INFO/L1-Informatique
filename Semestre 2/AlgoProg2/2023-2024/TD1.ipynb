```python
def est_rectangle(AB: float, AC: float, BC: float) -> bool:
    if AC*AC + AB*AB == BC*BC:
        return True
    if AB*AB + BC*BC == AC*AC:
        return True
    if AC*AC + BC*BC == AB*AB:
        return True
    return False
```

```python
assert est_rectangle(3,4,5) == True
assert est_rectangle(3,5,4) == True
assert est_rectangle(8,2,2) == False 
```

### Exercice 1.1.3

```python
def factorielle(n: int) -> int:
    '''Calculer la factorielle de n'''

    if n==0:
        return 1

    k=1
    for i in range(1,n+1):
        k= k*i
    return k
```

```python
from math import factorial
assert factorielle(5) == factorial(5)
assert factorielle(0) == factorial(0)
```

```python
def arrengement( n: int, k: int) -> float:

    assert k<=n and n>=0 and k>=0

    return factorielle(n)/factorielle(n-k)

```

```python
def combinaisons(n: int, k: int) -> float:

    assert k<=n and n>=0 and k>=0

    return arrengement(n,k)/factorielle(k)
    
```

```python
combinaisons(9,5)
```

```python
def tabup(t: list[int], n: int) -> bool:
    '''Verifie si les valeurs d'un tableau sont bien range dans l'ordre croissant'''
    for i in range(1,n-1):
        if t[i+1]< t[i]:
            return False
    
    return True    
```
