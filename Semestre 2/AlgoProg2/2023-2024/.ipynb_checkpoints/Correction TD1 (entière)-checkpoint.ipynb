```python
from math import factorial
#2
def factorielle(n:int)->int:
    i=1
    res=1
    while i <n:
        res=res+(res*i)
        i+=1
    return res
#3
assert factorielle(5)==factorial(5)
#4 
def arrangement(n,k:int)->int:
    res= factorielle(n)/factorielle(n-k)
    return res

```

```python
assert arrangement(3,2) == 6
```

```python
def com(n,k:int)->int:
    res=arrangement(n,k)/factorielle(k)
    return res
```

```python
assert com(5,2)== 10
```

```python
def moy(t : list[float], n : int) -> float:
    s= 0
    for i in range(len(t)):
        s=s+t[i]
    res=s/n
    return res

t=[float(input('entrez une val floatante=')) for i in range(4)]
print('la moy des val est', moy(t,4))
```

## 1.1.6. Exercice
#### Ecrire une fonctions tabup() qui vérifie si les n valeurs d’un tableau sont rangées par ordre croissant. Ces valeurs sont des entiers.


```python
def tabup(t : list[int])->bool:
    res=True
    for i in range(1,len(t)):
        if t[i]<t[i-1]:
            res=False
    return res

tabup([1,3,4])
```

### 1.1.9. Exercice 
#### Un palindrome est un mot, ou un groupe de mots, dont l’ordre des lettres reste le même qu’on le lise de la droite vers la gauche ou inversement. Des exemples bien connus sont “été”, “kayak”, “mon nom”, “élu par cette crapule”. Ce dernier permet d’illustrer qu’on ne tient pas compte en général des accents, trémas, cédilles, ni des espaces. Dans cet exercice :
1. un mot ou un groupe de mots est représenté par une chaîne de caractères (str),
2. ces caractères sont sans accent, tréma, ni cédille : “ete”
3. les espaces sont considérés comme des caractères : ainsi “elu par cette crapule” n’est pas un palindrome ici.






```python
#1 Ecrire une fonction qui teste si un argument est ou non un palindrome et retourne le booléen correspondant.

def pal(s : str)-> bool:
    res = False
    un = ' '
    d= ' '
    for i in range(0,len(s)//2+1):
        un += s[i]
    
    for m in range(len(s)-1, len(s)//2-1,-1):
        d += s[m]
    
    
    if d == un:
        res = True
    
    return res
```

```python
def palindrome(s : str, n : int) -> bool:
    for i in range(n):
        if s[i] != s[n-1-i]:
            return False
    return True
palindrome('radar',5)
```
